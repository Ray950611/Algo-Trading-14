This function is used to plot all stocks selected in a portfolio.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
%matplotlib inline
Dict_=np.load('data.npy').item()
L='fti tex tif sig hpq tjx sig fti sanm apc'.split(' ')
S='bvsn swn rad win nbr rad avp rrd wft nbr'.split(' ')
def plot(name,data,s,start,end,mid,action):
    plt.figure(figsize=(15,10))
    C=np.array(data[name][s])[start:end+1]
    l=np.array(data[name]['Date'])[start:end+1]
    for i in range(len(l)):
        l[i]=l[i][4:]
    plt.plot(l,C)
    plt.plot(np.array(data[name]['Date'])[mid][4:],np.array(data[name][s])[mid],'ro')
    plt.xlabel('Date')
    plt.ylabel('Close')
    plt.title(name.upper()+' '+action)
for i in set(L):
    plot(i,Dict_,'Close',44,62,53,'Long '+str(L.count(i)/float(len(L)))+' shares')
    plt.savefig('./figs/'+i+'_long_1217-1231.png')
for i in set(S):
    plot(i,Dict_,'Close',44,62,53,'Short'+str(S.count(i)/float(len(S)))+' shares')
    plt.savefig('./figs/'+i+'_short_1217-1231.png')
```

This function is used to plot a single stock's technical indicator information.

```python
def plot_single(name,data,start,end):
    
    plt.figure(figsize=(10,8))
    plt.xlabel('Dates')
    for s in ['Close','EMA']:
        C=np.array(data[name][s])[start:end+1]
        l=np.array(data[name]['Date'])[start:end+1]
        for i in range(len(l)):
            l[i]=l[i][4:]
        plt.plot(l,C,label=s)
    plt.xticks([0,10,20,30,40,50,60])
    plt.legend()
    plt.title(name.upper())
    plt.figure(figsize=(10,8))
    for s in ['RSI','SMI','SO']:
        C=np.array(data[name][s])[start:end+1]
        l=np.array(data[name]['Date'])[start:end+1]
        for i in range(len(l)):
            l[i]=l[i][4:]
        plt.plot(l,C,label=s)

    C=np.array(data[name]['MACD'])[start:end+1]
    l=np.array(data[name]['Date'])[start:end+1]
    for i in range(len(l)):
        l[i]=l[i][4:]
    plt.bar(l,C,label='MACD')
    plt.legend()
    plt.xticks([0,10,20,30,40,50,60])
    plt.title(name.upper())
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
%matplotlib inline
Dict_=np.load('data.npy').item()
plot_single('aapl',Dict_,0,62)
```
