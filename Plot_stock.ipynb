```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
%matplotlib inline
Dict_=np.load('data.npy').item()
L='kim ivz ctb dre anf pbi spxc nav nyt jbl'.split(' ')
def plot(name,data,s,start,end,mid):
    plt.figure(figsize=(15,10))
    C=np.array(data[name][s])[start:end+1]
    l=np.array(data[name]['Date'])[start:end+1]
    for i in range(len(l)):
        l[i]=l[i][4:]
    plt.plot(l,C)
    plt.plot(np.array(data[name]['Date'])[mid][4:],C[mid],'ro')
    plt.xlabel('Date')
    plt.ylabel('Close')
    plt.title(name)
for i in L:
    plot(i,Dict_,'Close',0,22,10)
```
