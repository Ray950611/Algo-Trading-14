```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
%matplotlib inline
Dict_=np.load('data.npy').item()
L='mac symc ctb pch vtr kim ivz ctb dre anf'.split(' ')
S='wft viav nbr chk dnr dnr esv chk do wft'.split(' ')
def plot(name,data,s,start,end,mid,action):
    plt.figure(figsize=(15,10))
    C=np.array(data[name][s])[start:end+1]
    l=np.array(data[name]['Date'])[start:end+1]
    for i in range(len(l)):
        l[i]=l[i][4:]
    plt.plot(l,C)
    plt.plot(np.array(data[name]['Date'])[mid][4:],C[mid],'ro')
    plt.xlabel('Date')
    plt.ylabel('Close')
    plt.title(name.upper()+' '+action)
for i in set(L):
    plot(i,Dict_,'Close',0,22,10,'Long '+str(L.count(i)/float(len(L)))+' shares')
    plt.savefig('./figs/'+i+'_long_1016-1031.png')
for i in S:
    plot(i,Dict_,'Close',0,22,10,'Short'+str(S.count(i)/float(len(S)))+' shares')
    plt.savefig('./figs/'+i+'_short_1016-1031.png')
```
